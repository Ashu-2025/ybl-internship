```python
# plot
plt.figure(figsize=[8,5])
grouped_data.plot(kind = 'bar',color='teal',edgecolor='blue')
plt.title('Average price of Food_Items')
plt.xlabel('Food_Item')
plt.ylabel('Average price[$]')
plt.show()
```

```python
grouped_data.values.mean()
```

* The above graph indicates the most expensive item in the menu is Pizza Margherita
* The above graph indicates that chapati is least expensive
* The average food item price is around 150.

```python
avg_order_price = df.groupby('Customer_Name')['Total'].sum() / df.groupby('Customer_Name')['Order_ID'].nunique()

avg_order_price.plot(kind='bar', figsize=(10,6))
plt.xlabel('Customer Name')
plt.ylabel('Average Order Price')
plt.title('Average Order Price Per Person')
plt.tight_layout()
plt.show()
```

```python
avg_order_price.mean()
```

* The above graph indicates the Vikram patel has spend more money than others
* The above graph indicates that sneha singh has spend least than others.
* The average person has ordered about 390.
